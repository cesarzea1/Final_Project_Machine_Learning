```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Read the specified number of rows from the CSV file
df = pd.read_csv(r'C:\Users\pola_\OneDrive\Documents\Cesar\Data-bootcamp-UoT\Lessons\week 23\final project\Final_Project_Machine_Learning\Resources\US_Accidents_March23.csv')

df
```

```python
# Print the information about the DataFrame including column names and data types
df.info()
```

```python
# Convert 'Start_Time' to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Sort by 'Start_Time'
df.sort_values(by='Start_Time', inplace=True)
```

```python
# Count the unique values in 'City'
city_counts = df['City'].value_counts()

# Calculate the total number of rows (to use for calculating proportions)
total_rows = len(df)

# Get the proportion of each 'City'
city_proportions = city_counts / total_rows

# proportions as percentages
city_percentages = city_proportions * 100

# print counts and proportions
print(city_counts)
print(city_proportions)
```

```python
# Convert city_counts to a DataFrame
city_counts_df = city_counts.reset_index()
city_counts_df.columns = ['City', 'Count']  

# Calculate proportions and add as a new column
city_counts_df['Proportion'] = city_counts_df['Count'] / city_counts_df['Count'].sum()

# Save the DataFrame to a CSV file
output_path = r'C:\Users\pola_\OneDrive\Documents\Cesar\Data-bootcamp-UoT\Lessons\week 23\final project\Final_Project_Machine_Learning\Resources\city_counts.csv'
city_counts_df.to_csv(output_path, index=False)
```

```python
# Sample 50% of the data randomly
sampled_df = df.sample(frac=0.5, random_state=1)

# Save the sampled DataFrame to a CSV file For the EDA in Tableau
sampled_output_path = r'C:\Users\pola_\OneDrive\Documents\Cesar\Data-bootcamp-UoT\Lessons\week 23\final project\Final_Project_Machine_Learning\Resources\sampled_US_Accidents.csv'
sampled_df.to_csv(sampled_output_path, index=False)
```

```python
# calculate the city counts and proportions for the sampled data
sampled_city_counts = sampled_df['City'].value_counts()

# Convert sampled city_counts to a DataFrame
sampled_city_counts_df = sampled_city_counts.reset_index()
sampled_city_counts_df.columns = ['City', 'Sampled_Count']  

# Calculate proportions for the sampled data and add as a new column
sampled_city_counts_df['Sampled_Proportion'] = sampled_city_counts_df['Sampled_Count'] / sampled_city_counts_df['Sampled_Count'].sum()

# Save the sampled city counts DataFrame to a CSV file
sampled_city_counts_output_path = r'C:\Users\pola_\OneDrive\Documents\Cesar\Data-bootcamp-UoT\Lessons\week 23\final project\Final_Project_Machine_Learning\Resources\sampled_city_counts.csv'
sampled_city_counts_df.to_csv(sampled_city_counts_output_path, index=False)
```

## Additional data cleaning.

```python
# Check for missing values in each column
missing_values = sampled_df.isnull().sum()

# Print the count of missing values for each column
print(missing_values)
```

```python
# Columns to drop
columns_to_drop = [
    'End_Lat', 'End_Lng', 'Street', 'City', 'Airport_Code', 
    'Weather_Timestamp', 'Wind_Chill(F)', 'Wind_Direction', 
    'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight',
    'Description', 'Timezone', 'Zipcode'
]

# Drop the columns
clean_sampled_df = sampled_df.drop(columns=columns_to_drop)

# Check the DataFrame to ensure the columns have been dropped
clean_sampled_df.info()
```

```python
# Check for missing values in each column
missing_values = clean_sampled_df.isnull().sum()

# Print the count of missing values for each column
print(missing_values)
```

```python
# Fill missing values for numerical columns with their mean
numerical_columns = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']
for column in numerical_columns:
    clean_sampled_df[column].fillna(clean_sampled_df[column].mean(), inplace=True)

# Fill missing values for 'Weather_Condition' and 'Astronomical_Twilight' with 'NA'
clean_sampled_df['Weather_Condition'].fillna('NA', inplace=True)
clean_sampled_df['Astronomical_Twilight'].fillna('NA', inplace=True)

# Verify the changes
print(clean_sampled_df.isnull().sum())
```
